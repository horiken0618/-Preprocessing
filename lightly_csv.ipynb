```python
out_list = []
for t in text:
    out_list.append(str(t))
out_text = "".join(out_list)
```

```python
import MeCab
mecabTagger = MeCab.Tagger()
noun_count = {}

for t in text:
    node = mecabTagger.parseToNode(str(t))
    while node:
        word = node.surface
        hinshi = node.feature.split(",")[0]
        if word in noun_count.keys() and hinshi == "名詞":
            noun_freq = noun_count[word]
            noun_count[word] = noun_freq + 1
        elif hinshi == "名詞":
            noun_count[word] = 1
        else:
            pass
        node = node.next
```

```python
noun_count = sorted(noun_count.items(), key=lambda x:x[1], reverse=True)
print(noun_count)
```

```python
# csvを軽くする関数
def lightly_data(file_name):
    # ライブラリの読み込み
    import csv
    import os
    import pandas as pd

    # オリジナルのデータを読み込み
    os.chdir("D:\じゃらん宿泊データ\original_data")
    csv_file_name = file_name
    columns = ["エリアコード","宿コード","データ取得日","チェックイン日","プランコード","価格","部屋情報","朝食の有無","夕食の有無","チェックイン時刻","チェックアウト時刻","プラン提供開始日","プラン提供終了日","イメージ画像URL","画像キャプション","残り部屋数"]
    df = pd.read_csv(csv_file_name,names=(columns))

    # "部屋情報","イメージ画像URL","画像キャプション"を削除する
    df_light = df.copy()
    df_light = df_light.drop(columns=["部屋情報","イメージ画像URL","画像キャプション"])
    df_light.head()
    os.chdir("D:\じゃらん宿泊データ\light_data")
    df_light.to_csv(csv_file_name)
    return print(csv_file_name)
```
